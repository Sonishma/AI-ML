1. Exercise on Functions

Question 1-Unit Conversion Program
Create a Python program that converts between units.

Question 2-Mathematical Operations on a List
Create a program that performs operations on a list of numbers.


```python
#Question 1
def convert_units(conversion_type, value):
    """
    Converts a value between supported units.

    Parameters:
        conversion_type (str): Type of conversion to perform.
            Supported options:
            - 'm_to_ft'
            - 'ft_to_m'
            - 'kg_to_lbs'
            - 'lbs_to_kg'
            - 'l_to_gal'
            - 'gal_to_l'
        value (float): Numeric value to convert.

    Returns:
        float: Converted value.

    Raises:
        ValueError: If the conversion type is not supported.
    """
    if conversion_type == "m_to_ft":
        return value * 3.28084
    elif conversion_type == "ft_to_m":
        return value / 3.28084
    elif conversion_type == "kg_to_lbs":
        return value * 2.20462
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462
    elif conversion_type == "l_to_gal":
        return value * 0.264172
    elif conversion_type == "gal_to_l":
        return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type.")


def unit_conversion_program():
    print("Choose conversion type:")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")

    try:
        category = input("Enter your choice (1/2/3): ").strip()

        if category == "1":
            print("\nLength conversions:")
            print("a. meters to feet")
            print("b. feet to meters")
            sub = input("Choose option (a/b): ").strip().lower()

            if sub == "a":
                value = float(input("Enter value in meters: "))
                result = convert_units("m_to_ft", value)
                print(f"{value} meters = {result:.4f} feet")
            elif sub == "b":
                value = float(input("Enter value in feet: "))
                result = convert_units("ft_to_m", value)
                print(f"{value} feet = {result:.4f} meters")
            else:
                print("Invalid length conversion option.")

        elif category == "2":
            print("\nWeight conversions:")
            print("a. kilograms to pounds")
            print("b. pounds to kilograms")
            sub = input("Choose option (a/b): ").strip().lower()

            if sub == "a":
                value = float(input("Enter value in kilograms: "))
                result = convert_units("kg_to_lbs", value)
                print(f"{value} kilograms = {result:.4f} pounds")
            elif sub == "b":
                value = float(input("Enter value in pounds: "))
                result = convert_units("lbs_to_kg", value)
                print(f"{value} pounds = {result:.4f} kilograms")
            else:
                print("Invalid weight conversion option.")

        elif category == "3":
            print("\nVolume conversions:")
            print("a. liters to gallons")
            print("b. gallons to liters")
            sub = input("Choose option (a/b): ").strip().lower()

            if sub == "a":
                value = float(input("Enter value in liters: "))
                result = convert_units("l_to_gal", value)
                print(f"{value} liters = {result:.4f} gallons")
            elif sub == "b":
                value = float(input("Enter value in gallons: "))
                result = convert_units("gal_to_l", value)
                print(f"{value} gallons = {result:.4f} liters")
            else:
                print("Invalid volume conversion option.")

        else:
            print("Invalid category selected.")

    except ValueError:
        print("Invalid input. Please enter numeric values where required.")

unit_conversion_program()
```

```python
#Question 2
def find_sum(numbers):
    """
    Returns the sum of a list of numbers.

    Parameters:
        numbers (list): List of numeric values.

    Returns:
        float: Sum of numbers.
    """
    return sum(numbers)


def find_average(numbers):
    """
    Returns the average of a list of numbers.

    Parameters:
        numbers (list): List of numeric values.

    Returns:
        float: Average of numbers.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Returns the maximum value from a list of numbers.

    Parameters:
        numbers (list): List of numeric values.

    Returns:
        float: Maximum value.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Returns the minimum value from a list of numbers.

    Parameters:
        numbers (list): List of numeric values.

    Returns:
        float: Minimum value.
    """
    return min(numbers)


def math_operations_program():
    print("Choose operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = input("Enter your choice (1/2/3/4): ").strip()
        user_input = input("Enter numbers separated by spaces: ").strip()

        if not user_input:
            raise ValueError("Empty input is not allowed.")

        numbers = [float(x) for x in user_input.split()]

        if len(numbers) == 0:
            raise ValueError("List cannot be empty.")

        if choice == "1":
            print("Sum =", find_sum(numbers))
        elif choice == "2":
            print("Average =", find_average(numbers))
        elif choice == "3":
            print("Maximum =", find_maximum(numbers))
        elif choice == "4":
            print("Minimum =", find_minimum(numbers))
        else:
            print("Invalid operation selected.")

    except ValueError:
        print("Invalid input. Please enter valid numbers.")

math_operations_program()
```

2. Exercise on List Manipulation

Question 1-Extract Every Other Element

Question 2-Slice a Sublist

Question 3-Reverse a List Using Slicing

Question 4-Remove First and Last Elements

Question 5-Get First n Elements

Question 6-Get Last n Elements

Question 7-Reverse Skip Elements


```python
#Question 1
def extract_every_other(lst):
    """
    Returns every other element from the list, starting from the first element.

    Parameters:
        lst (list): Input list.

    Returns:
        list: New list containing every other element.
    """
    return lst[::2]


print(extract_every_other([1, 2, 3, 4, 5, 6]))
```

```python
#Question 2
def get_sublist(lst, start, end):
    """
    Returns a sublist from start index to end index.

    Parameters:
        lst (list): Input list.
        start (int): Starting index.
        end (int): Ending index.

    Returns:
        list: Extracted sublist.
    """
    return lst[start:end + 1]


print(get_sublist([1, 2, 3, 4, 5, 6], 2, 4))
```

```python
#Question 3
def reverse_list(lst):
    """
    Returns the reversed version of the list using slicing.

    Parameters:
        lst (list): Input list.

    Returns:
        list: Reversed list.
    """
    return lst[::-1]


print(reverse_list([1, 2, 3, 4, 5]))
```

```python
#Question 4
def remove_first_last(lst):
    """
    Removes the first and last elements of a list.

    Parameters:
        lst (list): Input list.

    Returns:
        list: List without the first and last elements.
    """
    return lst[1:-1]


print(remove_first_last([1, 2, 3, 4, 5]))
```

```python
#Question 5
def get_first_n(lst, n):
    """
    Returns the first n elements of the list.

    Parameters:
        lst (list): Input list.
        n (int): Number of elements to return.

    Returns:
        list: First n elements.
    """
    return lst[:n]


print(get_first_n([1, 2, 3, 4, 5], 3))
```

```python
#Question 6
def get_last_n(lst, n):
    """
    Returns the last n elements of the list.

    Parameters:
        lst (list): Input list.
        n (int): Number of elements to return.

    Returns:
        list: Last n elements.
    """
    return lst[-n:] if n != 0 else []


print(get_last_n([1, 2, 3, 4, 5], 2))
```

```python
#Question 7
def reverse_skip(lst):
    """
    Returns every second element starting from the second-to-last element,
    moving backward.

    Parameters:
        lst (list): Input list.

    Returns:
        list: Extracted elements in reverse-skip order.
    """
    return lst[-2::-2]


print(reverse_skip([1, 2, 3, 4, 5, 6]))
```

3. Exercise on Nested Lists

Question 1-Flatten a Nested List

Question 2-Access Nested Element

Question 3-Sum of All Elements in Nested List

Question 4-Remove Element from Nested List

Question 5-Find Maximum Element in Nested List

Question 6-Count Occurrences in Nested List

Question 7-Deep Flatten

Question 8-Average of Nested List


```python
#Question 1
def flatten(lst):
    """
    Flattens a nested list by one or more levels recursively.

    Parameters:
        lst (list): Nested list.

    Returns:
        list: Flattened list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result


print(flatten([[1, 2], [3, 4], [5]]))
```

```python
#Question 2
def access_nested_element(lst, indices):
    """
    Accesses an element from a nested list using a list of indices.

    Parameters:
        lst (list): Nested list.
        indices (list): List of indices showing the path.

    Returns:
        Any: The element at the specified nested position.
    """
    element = lst
    for index in indices:
        element = element[index]
    return element


lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(access_nested_element(lst, [1, 2]))
```

```python
#Question 3
def sum_nested(lst):
    """
    Returns the sum of all numeric elements in a nested list.

    Parameters:
        lst (list): Nested list.

    Returns:
        int/float: Sum of all elements.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total


print(sum_nested([[1, 2], [3, [4, 5]], 6]))
```

```python
#Question 4
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
        lst (list): Nested list.
        elem (Any): Element to remove.

    Returns:
        list: Modified nested list with the element removed.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            cleaned = remove_element(item, elem)
            result.append(cleaned)
        else:
            if item != elem:
                result.append(item)
    return result


print(remove_element([[1, 2], [3, 2], [4, 5]], 2))
```

```python
#Question 5
def find_max(lst):
    """
    Finds the maximum element in a nested list.

    Parameters:
        lst (list): Nested list.

    Returns:
        int/float: Maximum element.
    """
    flattened = flatten(lst)
    return max(flattened)


print(find_max([[1, 2], [3, [4, 5]], 6]))
```

```python
#Question 6
def count_occurrences(lst, elem):
    """
    Counts how many times an element appears in a nested list.

    Parameters:
        lst (list): Nested list.
        elem (Any): Element to count.

    Returns:
        int: Number of occurrences.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count


print(count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
```

```python
#Question 7
def deep_flatten(lst):
    """
    Deeply flattens a nested list regardless of depth.

    Parameters:
        lst (list): Deeply nested list.

    Returns:
        list: Fully flattened list.
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result


print(deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
```

```python
#Question 8
def average_nested(lst):
    """
    Calculates the average of all numbers in a nested list.

    Parameters:
        lst (list): Nested list.

    Returns:
        float: Average of all elements.
    """
    flattened = deep_flatten(lst)
    return sum(flattened) / len(flattened)


print(average_nested([[1, 2], [3, 4], [5, 6]]))
```
